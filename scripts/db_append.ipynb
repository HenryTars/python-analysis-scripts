```python
import sqlite3
import pandas as pd
from tqdm import tqdm  # for progress bar

# === CONFIGURATION ===
csv_file = "E:/ACVWS_ALL_AC_ENTRIES_ACRJRNAL_MAY_2025.csv"
db_file = "E:/cbs.db"
table_name = "ACVWS_ALL_AC_ENTRIES_ACRJRNAL_2025"
chunksize = 100_000

# === IMPORT PROCESS ===
try:
    # Count total rows for progress (excluding header)
    total_rows = sum(1 for _ in open(csv_file, encoding="utf-8")) - 1
    total_chunks = (total_rows // chunksize) + 1

    print(f"ðŸ“¦ Starting import: {total_rows:,} rows, {total_chunks} chunks")

    conn = sqlite3.connect(db_file)

    # Read CSV in chunks and append to SQLite
    chunk_iter = pd.read_csv(csv_file, chunksize=chunksize)
    for i, chunk in enumerate(tqdm(chunk_iter, total=total_chunks, desc="Importing chunks")):
        chunk.to_sql(table_name, conn, if_exists='append', index=False)
    
    conn.close()
    print("âœ… All chunks imported successfully!")

except Exception as e:
    print(f"âŒ Import failed: {e}")
```
